```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = 1.5e-3
a = 0.5
i_0 = 0.0
t0 = 0.0

t1 = 20.0
n = 5000.0
dt = (t1-t0)/n
L = np.zeros(5000) #5000 zeros in numpoy in array form

for i in np.arange(1,5000):
    L[i] = L[i-1] + (w-a*L[i-1])*dt
```

```python
%matplotlib notebook
fig, ax = plt.subplots()

ax.plot(L)
```
