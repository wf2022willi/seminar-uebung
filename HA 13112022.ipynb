```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,num=10000)

def f1(x):
    return x**1
def f2(x):
    return x**2
def f3(x):
    return x**3
def f4(x):
    return x**4

```
